## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. setup plot
fig,ax = plt.subplots(figsize = (10,10)) # Width and hight

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title="simple plot",
      xlabel="x-axis",
      ylabel="y-axis")
# 5. save & show
fig.savefig("../img/sample-plot.png")
```

```python
## Making Figures with Numpy arrays
import numpy as np
# Create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
# Plot the data , Create a lineplot
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Use same data to make a scatterplot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Another scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
```

```python
# Making a plot from dictionary 
nut_butter_prices={"Almond Butter":10,
                  "Peanut Butter":8,
                  "Cashew Butter":12}
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Dan's Nut Butter store",
      ylabel="Prices($)")
```

```python
# Horizontal bar graph
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
```

```python
# Make some data for histogram
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x)
```

```python
# Two ways for subplot
# Option 1
fig,((ax1,ax2),(ax3,ax4))= plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option Two
fig,ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5));
# Plot to eaxh different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
car_sales["Price"] = car_sales["Price"].str.replace("[\$\,\.]",'');
car_sales
```

```python
# Remove extra two extra zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2023",periods=len(car_sales));
car_sales
```

```python
# Clean the "Price" column by removing non-numeric characters and converting to integers
car_sales["Price"] = car_sales["Price"].str.replace('[\$,]', '', regex=True).astype(float)

# Cumulative sum of the cleaned "Price" column
car_sales["Total sales"] = car_sales["Price"].cumsum()

car_sales
```

```python
# lets plot the total Sales
car_sales.plot(x="Sale Date",y="Total sales")
```

```python
# Changing type from string to intiger
car_sales["Price"] = car_sales["Price"].astype(int)
# Plot a scatter plot with price column as numeric
car_sales.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

### Which one should we use?

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
over_50.plot(kind='scatter',
            x = 'age',
            y = 'chol',
            c = 'target');
```

```python
# 00 method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = 'scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);
ax.set_xlim([45,100])
```

```python
# OO method from scratch
fig , ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

# custamize the plot
ax.set(title = "Heart Disease",
       xlabel = "age",
       ylabel = "Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
```

```python
over_50.head()
```

```python
import matplotlib.pyplot as plt
# subplot of chol, age , thalch
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10));

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]);

# customize ax0
ax0.set(title="Heart disease and cholestrol level",
        xlabel="Age",
        ylabel="Cholesterol");

# Add legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# Add meanLine
ax0.axhline(y=over_50["chol"].mean(),
          linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# customize ax1
ax1.set(title="Heart disease and Max heart rate",
        xlabel="Age",
        ylabel="Max Heart Rate");

# Add a legend
ax1.legend(*scatter.legend_elements(),title="Target");

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add Title to the figure
fig.suptitle("Heart disease Analysis",
            fontsize=16,
            fontweight="bold");
```

```python
# subplot of chol, age , thalch
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10));

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter");

# customize ax0
ax0.set(title="Heart disease and cholestrol level",
        xlabel="Age",
        ylabel="Cholesterol");

ax0.set_xlim(50,80)
ax0.set_ylim(100,600)

# Add legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target");

# Add meanLine
ax0.axhline(y=over_50["chol"].mean(),
          linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="plasma")

# customize ax1
ax1.set(title="Heart disease and Max heart rate",
        xlabel="Age",
        ylabel="Max Heart Rate");
ax1.set_xlim(50,80)
ax1.set_ylim(60,200)

# Add a legend
ax1.legend(*scatter.legend_elements(),title="Target");

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

# Add Title to the figure
fig.suptitle("Heart disease Analysis",
            fontsize=16,
            fontweight="bold")
```
